### 1.3 - Matplotlib 
A very powerful toolbox for visualization in Python is Matplotlib that features possibilities for plotting in both 2 and 3 dimensions. It works perfectly in combination with Numpy to provide the underlying data.

#### 1.3.1 - Line Plots
Creat a figure object (figure) with an empty axis. Then plot the graph of the function $f(x)=\sin(x)$ on the interval $[0,2\pi]$ into the diagram.

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck

# YOUR CODE HERE
x = np.linspace(0, 2*np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x/np.pi, np.sin(x))

ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=0.5))

ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(x)$')
ax.set_title(r'$f(x)=\sin(x)$')

# display the plot
ax.grid()
plt.show()
````

#### 1.3.2 - Surface plots
Use a meshgrid over $[-2,2]\times [-2,2]$ to visualize the isosurface given by the graph of $f(x,y) = x^2+y^2$.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# YOUR CODE HERE
plt.rcParams["figure.figsize"] = [5, 5]
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
x, y = np.meshgrid(x, y)
z = x**2 + y**2
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, cmap=plt.cm.coolwarm)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$f(x,y)$')
ax.set_title(r'$f(x,y)=x^2+y^2$')
plt.show()
````
